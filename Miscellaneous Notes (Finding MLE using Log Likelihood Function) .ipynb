### Observe 3 failures and 1 success  then L(pi|x)=(1-pi)^3*^3*pi.1  

```r
library(dplyr)
pi= seq(0.01,0.99, length=99) #list out all 99 values 
pi #these are the values used on the horizontal axis
lnlik=log(pi) + 3*log(1-pi) # because ln(0) is infinity and if pi=1, 3*log(1-pi) would also equal infinity. 
                            # So pi should be between 0 and 1. 
                            #Each value of pi is passed thru lnlik.
                            # MLE must be one of the values of pi; otherwise that value will not be observed.  

# find approximate max(lik) in R
topl=max(lnlik) #This is the highest output of lnlik which is -2.24934057847523
topl 
newx2 <- pi[which(lnlik == topl)]
newx2 #This is the parameter that gives the highest value. Pi=0.25


#Table format
temp <- data.frame(pi,lnlik) 
temp
temp[which.max(temp$lnlik),] #This selects the pi with the greatest lnlik value. 


#Graph Log-likelihood 
plot(pi, lnlik, xlab="pi", ylab="Log[Likelihood]", type="l")
abline(v=newx2)
abline(h=topl) 
```

### Show that the ML estimate of $\pi$ is 0.50

```r
# The log[likelihood] maximizes at the same value of pi as the likelihood 
# Here log[L(pi|x)]= ln(pi)+ln(1-pi)
# Note that the log function in R uses base e
pi= seq(0.01,0.99, length=99) # because ln(0) is infinity
lnlik=log(pi) + log(1-pi)  #because ln(0) is infinity and if pi=1, log(1-pi) would also equal infinity. 
                            # So pi should be between 0 and 1. 
                            #Each value of pi is passed thru lnlik.
                            # MLE must be one of the values of pi; otherwise that value will not be observed.  
topl=max(lnlik)
topl #this is the greatest value for the log-likelihood function; not the parameter that gives the highest value

# find approximate max(lik) in R
newx2 <- pi[which(lnlik == topl)]
newx2 #This is the parameter that gives the highest value 


#Table format 
temp <- data.frame(pi,lnlik) 
temp
temp[which.max(temp$lnlik),] #This selects the pi with the greatest lnlik value. 

#Graph of Log-likelihood
plot(pi, lnlik, xlab="pi", ylab="Log[Likelihood]", type="l")
abline(v=newx2) #parameter that gives the highest output
abline(h=topl) #highest output of the log-likelihood function 
```
